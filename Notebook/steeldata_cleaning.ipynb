```python
#I did not preform the correct average for the data.  I am know I will probably delete these rows in the future for analysis.
#However, I want to have some value in there and I still want the average of the last known value before missing, with the last value.
#So, here we go again, but fixing via rows.  I have nice looking columns now.
cols_to_fix = ["temperature_2m (C)", "relative_humidity_2m (%)", "dew_point_2m (C)", "precipitation (mm)"]

# Get the last available row (23:00) and next row (00:00)
last_valid_values = merged_df.loc[merged_df["date"] == "2018-12-31 23:00:00", cols_to_fix].iloc[0]
next_valid_values = merged_df.loc[merged_df["date"] == "2018-12-31 00:00:00", cols_to_fix].iloc[0]

# Overwrite the incorrect values by averaging the last and next row
for col in cols_to_fix:
    merged_df.loc[merged_df["date"].between("2018-12-31 23:15:00", "2018-12-31 23:45:00"), col] = \
        (last_valid_values[col] + next_valid_values[col]) / 2

```
