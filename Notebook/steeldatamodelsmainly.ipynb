```python
#Model checks on ones that showed significance and comparisons 

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Load dataframe
try:
    df = pd.read_parquet('df_no_outliers.parquet')
    df['date'] = pd.to_datetime(df['date'])
except FileNotFoundError:
    print("Error: 'df_no_outliers.parquet' not found.")
    exit()

# Verify columns
required_cols = ['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'relative_humidity_2m (%)', 'temperature_2m (°C)']
if not all(col in df.columns for col in required_cols):
    print("Error: Missing columns:", [col for col in required_cols if col not in df.columns])
    exit()

# Feature humidity  think it has a impact
df['log_humidity'] = np.log1p(df['relative_humidity_2m (%)'])
df['humidity_binned'] = pd.cut(df['relative_humidity_2m (%)'], bins=[0, 70, 85, 100], labels=['Low', 'Medium', 'High'])
le = LabelEncoder()
df['humidity_binned'] = le.fit_transform(df['humidity_binned'])

# Features and target
X_rf = df[['Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)']]
X_xgb = df[['Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)', 'humidity_binned']]
X_no_humidity = df[['Lagging_Current_Reactive.Power_kVarh', 'temperature_2m (°C)']]
y = df['Usage_kWh']
X_rf_train, X_rf_test, y_train, y_test = train_test_split(X_rf, y, test_size=0.2, random_state=42)
X_xgb_train, X_xgb_test, y_xgb_train, y_xgb_test = train_test_split(X_xgb, y, test_size=0.2, random_state=42)
X_no_humidity_train, X_no_humidity_test, y_no_humidity_train, y_no_humidity_test = train_test_split(X_no_humidity, y, test_size=0.2, random_state=42)

# Random Forest (Humidity)
model_rf = RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42)
model_rf.fit(X_rf_train, y_train)
rf_r2 = model_rf.score(X_rf_test, y_test)
rf_mse = mean_squared_error(y_test, model_rf.predict(X_rf_test))
print(f'Random Forest (Humidity) R²: {rf_r2:.4f}, MSE: {rf_mse:.2f}')
print('Feature Importance:\n', pd.Series(model_rf.feature_importances_, index=X_rf.columns).sort_values(ascending=False))

# Random Forest (No Humidity)
model_rf_no_humidity = RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42)
model_rf_no_humidity.fit(X_no_humidity_train, y_no_humidity_train)
rf_no_humidity_r2 = model_rf_no_humidity.score(X_no_humidity_test, y_no_humidity_test)
rf_no_humidity_mse = mean_squared_error(y_no_humidity_test, model_rf_no_humidity.predict(X_no_humidity_test))
print(f'Random Forest (No Humidity) R²: {rf_no_humidity_r2:.4f}, MSE: {rf_no_humidity_mse:.2f}')

# XGBoost (Humidity)
model_xgb = XGBRegressor(n_estimators=50, max_depth=5, random_state=42)
model_xgb.fit(X_xgb_train, y_xgb_train)
xgb_r2 = model_xgb.score(X_xgb_test, y_xgb_test)
xgb_mse = mean_squared_error(y_xgb_test, model_xgb.predict(X_xgb_test))
print(f'XGBoost (Humidity) R²: {xgb_r2:.4f}, MSE: {xgb_mse:.2f}')
print('Feature Importance:\n', pd.Series(model_xgb.feature_importances_, index=X_xgb.columns).sort_values(ascending=False))

# XGBoost (No Humidity)
model_xgb_no_humidity = XGBRegressor(n_estimators=50, max_depth=5, random_state=42)
model_xgb_no_humidity.fit(X_no_humidity_train, y_no_humidity_train)
xgb_no_humidity_r2 = model_xgb_no_humidity.score(X_no_humidity_test, y_no_humidity_test)
xgb_no_humidity_mse = mean_squared_error(y_no_humidity_test, model_xgb_no_humidity.predict(X_no_humidity_test))
print(f'XGBoost (No Humidity) R²: {xgb_no_humidity_r2:.4f}, MSE: {xgb_no_humidity_mse:.2f}')

# OLS (Humidity)
X_ols = sm.add_constant(X_xgb)
model_ols = sm.OLS(y, X_ols).fit()
ols_r2 = model_ols.rsquared
ols_mse = mean_squared_error(y, model_ols.predict(X_ols))
print(f'OLS (Humidity) R²: {ols_r2:.4f}, MSE: {ols_mse:.2f}')

# SARIMAX (Humidity)
try:
    df_sarimax = df[['date', 'Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)']].set_index('date')
    df_sarimax = df_sarimax.sort_index().loc[~df_sarimax.index.duplicated(keep='first')]
    full_index = pd.date_range(start=df_sarimax.index.min(), end=df_sarimax.index.max(), freq='h')
    df_sarimax = df_sarimax.resample('h').mean().reindex(full_index)
    df_sarimax[['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)']] = \
        df_sarimax[['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)']].astype(float)
    df_sarimax = df_sarimax.infer_objects(copy=False).interpolate(method='time')
    df_sarimax.index.freq = 'h'
    y_time = df_sarimax['Usage_kWh']
    exog = df_sarimax[['Lagging_Current_Reactive.Power_kVarh', 'log_humidity', 'temperature_2m (°C)']]
    train_size = int(0.8 * len(y_time))
    y_train_time, y_test_time = y_time[:train_size], y_time[train_size:]
    exog_train, exog_test = exog[:train_size], exog[train_size:]
    model_sarimax = SARIMAX(y_train_time, exog=exog_train, order=(1,1,1), seasonal_order=(1,0,1,24))
    results_sarimax = model_sarimax.fit(disp=False, maxiter=200)
    forecast = results_sarimax.forecast(steps=len(y_test_time), exog=exog_test)
    sarimax_mse = mean_squared_error(y_test_time, forecast)
    sarimax_r2 = r2_score(y_test_time, forecast)
    print(f'SARIMAX (Humidity) R²: {sarimax_r2:.4f}, MSE: {sarimax_mse:.2f}')
except Exception as e:
    print(f'SARIMAX Error: {e}')
    sarimax_r2, sarimax_mse = None, None

# Model comparison
results = {
    'Model': ['Random Forest (Humidity)', 'Random Forest (No Humidity)', 'XGBoost (Humidity)', 'XGBoost (No Humidity)', 
              'OLS (Humidity)', 'SARIMAX (Humidity)', 'OLS (No Humidity)', 'VECM'],
    'R²': [rf_r2, rf_no_humidity_r2, xgb_r2, xgb_no_humidity_r2, ols_r2, sarimax_r2, 0.644, None],
    'MSE': [rf_mse, rf_no_humidity_mse, xgb_mse, xgb_no_humidity_mse, ols_mse, 
            sarimax_mse if sarimax_mse is not None else 'Error', None, 1815.63]
}
print(pd.DataFrame(results))

# Validation checks
# Cross-validation
cv_scores_rf = cross_val_score(RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42), X_rf, y, cv=5, scoring='r2')
print(f'Random Forest CV R²: Mean = {cv_scores_rf.mean():.4f}, Std = {cv_scores_rf.std():.4f}')
cv_scores_xgb = cross_val_score(XGBRegressor(n_estimators=50, max_depth=5, random_state=42), X_xgb, y, cv=5, scoring='r2')
print(f'XGBoost CV R²: Mean = {cv_scores_xgb.mean():.4f}, Std = {cv_scores_xgb.std():.4f}')

# Residual plots
rf_pred = model_rf.predict(X_rf_test)
rf_resid = y_test - rf_pred
plt.scatter(rf_pred, rf_resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Usage_kWh')
plt.ylabel('Residuals')
plt.title('Random Forest Residuals')
plt.show()
xgb_pred = model_xgb.predict(X_xgb_test)
xgb_resid = y_xgb_test - xgb_pred
plt.scatter(xgb_pred, xgb_resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Usage_kWh')
plt.ylabel('Residuals')
plt.title('XGBoost Residuals')
plt.show()

# Save to GitHub
# git add models.ipynb; git commit -m "Validated Random Forest and XGBoost, retained all models for portfolio"; git push origin main
```

```python
#create the missing features and rerun the XGBoost model. 
#Since df_no_outliers is derived from steel_weather_data.csv with zeros imputed and outliers removed
#need to ensure all features are generated correctly.
#Step 1: Create Time-Based and Lag Features


import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error

# Load DataFrame
df = pd.read_csv('steel_weather_data.csv')

# Impute zeros
df['CO2_imputed'] = df.groupby('humidity_binned')['CO2(tCO2)'].transform(
    lambda x: x.replace(0, x[x != 0].mean() if x[x != 0].mean() else x)
)

# Remove outliers
Q1 = df['CO2_imputed'].quantile(0.25)
Q3 = df['CO2_imputed'].quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[(df['CO2_imputed'] >= Q1 - 1.5 * IQR) & (df['CO2_imputed'] <= Q3 + 1.5 * IQR)].copy()

# Create time-based features
df_no_outliers['date_time'] = pd.to_datetime(df_no_outliers['date_time'])
df_no_outliers['hour_of_day'] = df_no_outliers['date_time'].dt.hour
df_no_outliers['is_weekend'] = df_no_outliers['day_of_week'].isin([5, 6]).astype(int)

# Create lag features
for lag in range(1, 5):
    df_no_outliers[f'CO2_lag_{lag}'] = df_no_outliers['CO2_imputed'].shift(lag)
    df_no_outliers[f'kWh_lag_{lag}'] = df_no_outliers['Usage_kWh'].shift(lag)

# Drop rows with NaN lags
df_no_outliers = df_no_outliers.dropna()

# Verify features
print("Columns after feature creation:", df_no_outliers.columns.tolist())
```

```python
#Now that the new features (varibles) are shown as columns
#Lets run the XGBOOST with them. Remember, it will be hard to get a better R2 or .998
# Define features
features = [
    'temperature_2m (°C)', 'Usage_kWh', 'relative_humidity_2m (%)',
    'hour_of_day', 'day_of_week', 'is_weekend',
    'CO2_lag_1', 'CO2_lag_2', 'CO2_lag_3', 'CO2_lag_4',
    'kWh_lag_1', 'kWh_lag_2', 'kWh_lag_3', 'kWh_lag_4'
]

# Prepare data
X = df_no_outliers[features]
y = df_no_outliers['CO2_imputed']
X = pd.get_dummies(X, columns=['day_of_week'], drop_first=True)  # Encode categorical

# Train XGBoost
model = XGBRegressor()
model.fit(X, y)
y_pred = model.predict(X)
print(f"XGBoost with Lags R²: {r2_score(y, y_pred):.3f}")
print(f"XGBoost with Lags MSE: {mean_squared_error(y, y_pred):.3f}")

# Feature importance
importances = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
print(importances.sort_values(by='Importance', ascending=False))

# Save DataFrame
df_no_outliers.to_parquet('steel_data.parquet')
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r'C:/Users/srsch/OneDrive/Desktop/steel_weather_data_restored_v2.csv')
print("Original shape:", df.shape)

# Recompute humidity_binned
df['humidity_binned'] = pd.qcut(df['relative_humidity_2m (%)'], q=3, labels=['Low', 'Medium', 'High'])

# Recompute CO2_imputed (use global mean for zeros to increase variance)
global_mean = df[df['CO2(tCO2)'] != 0]['CO2(tCO2)'].mean()
df['CO2_imputed'] = df['CO2(tCO2)'].replace(0, global_mean)

# Recalculate lags for CO2_imputed
for lag in range(1, 5):
    df[f'CO2_lag_{lag}'] = df['CO2_imputed'].shift(lag)

# Drop rows with NaN lags
df = df.dropna()

# Verify
print("New shape:", df.shape)
print("CO2_imputed summary:")
print(df['CO2_imputed'].describe())
print("CO2_imputed variance by humidity bin:")
print(df.groupby('humidity_binned')['CO2_imputed'].var())
print("humidity_binned value counts:")
print(df['humidity_binned'].value_counts())
print("relative_humidity_2m (%) range by bin:")
print(df.groupby('humidity_binned')['relative_humidity_2m (%)'].agg(['min', 'max']))
print("CO2_imputed medians:")
print(df.groupby('humidity_binned')['CO2_imputed'].median())

# Save updated dataset
output_csv = r'C:/Users/srsch/OneDrive/Desktop/steel_weather_data_restored_v3.csv'
df.to_csv(output_csv, index=False)
print(f"Updated dataset saved to {output_csv}")
```

```python
#Final XGBoost Model (v27, still trying to achieve my previous modell of R² = 0.998)
#To hit Test R² = 0.998 while maintaining log_kWh ~96.50%, will reintroduce CO2_lag_1 with controlled weight
#using a slightly broader feature set, and optimize hyperparameters with early stopping.


from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r'C:/Users/srsch/OneDrive/Desktop/steel_weather_data_restored_v7.csv')
print("Dataset shape:", df.shape)
print("Low humidity rows:", len(df[df['humidity_binned'] == 'Low']))

# Create log_kWh feature
df['log_kWh'] = np.log1p(df['Usage_kWh'])

# Define features (include CO2_lag_1 with minimal others)
features = ['log_kWh', 'CO2_lag_1', 'is_weekend', 'day_of_week_Tuesday']
print("Features used:", features)

# Low Humidity subset
df_low = df[df['humidity_binned'] == 'Low'].copy()
X_low = df_low[features]
y_low = df_low['CO2_imputed']

# Check feature statistics
print("Feature variances:", X_low.var())

# Scale features
scaler = StandardScaler()
X_low_scaled = scaler.fit_transform(X_low)
X_low_scaled = pd.DataFrame(X_low_scaled, columns=features)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_low_scaled, y_low, test_size=0.2, random_state=42)

# Train model (optimized with early stopping)
model_low = XGBRegressor(
    n_estimators=3000, max_depth=3, learning_rate=0.003, 
    min_child_weight=1, subsample=0.9, colsample_bytree=0.6, random_state=42,
    early_stopping_rounds=100, eval_metric='rmse'
)
model_low.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)

# Evaluate
y_train_pred = model_low.predict(X_train)
y_test_pred = model_low.predict(X_test)
print("Low Humidity Train R²:", r2_score(y_train, y_train_pred))
print("Low Humidity Test R²:", r2_score(y_test, y_test_pred))
print("Low Humidity Train MSE:", mean_squared_error(y_train, y_train_pred))
print("Low Humidity Test MSE:", mean_squared_error(y_test, y_test_pred))

# Save feature importance
importances = pd.DataFrame({'Feature': features, 'Importance': model_low.feature_importances_})
print("Feature Importance:")
print(importances.sort_values('Importance', ascending=False))
importances.to_csv(r'C:/Users/srsch/OneDrive/Desktop/feature_importance_low_xgboost_v27.csv', index=False)
print("Feature importance saved to feature_importance_low_xgboost_v27.csv")

```

```python
# Alternative flowchart using matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 14))

# Function to create a node
def create_node(x, y, width, height, text, color='lightblue'):
    node = patches.Rectangle((x, y), width, height, facecolor=color, edgecolor='black', alpha=0.7)
    ax.add_patch(node)
    
    # Handle multi-line text
    lines = text.split('\n')
    line_height = height / (len(lines) + 1)
    
    for i, line in enumerate(lines):
        ax.text(x + width/2, y + height - (i+1)*line_height, 
                line, ha='center', va='center', fontsize=9)
    
    return (x + width/2, y)  # Return bottom center point for connections

# Function to draw an arrow
def draw_arrow(start, end, label=None):
    ax.arrow(start[0], start[1], end[0]-start[0], end[1]-start[1], 
             head_width=0.1, head_length=0.1, fc='black', ec='black', length_includes_head=True)
    
    if label:
        mid_x = (start[0] + end[0]) / 2
        mid_y = (start[1] + end[1]) / 2
        ax.text(mid_x, mid_y, label, ha='center', va='center', fontsize=8, 
                bbox=dict(facecolor='white', alpha=0.7, boxstyle='round,pad=0.2'))

# Create nodes
node_width = 3
node_height = 0.8

# Position nodes
A = create_node(3, 13, node_width, node_height, 'MODEL(s) CREATIONS')
B = create_node(3, 12, node_width, node_height, 'Data Collection\nSteel Industry (Kaggle) + Weather (Gwangyang)')
C = create_node(3, 10.5, node_width, node_height, 'Data Preprocessing\nMerged, Imputed CO2 Zeros (Bin Medians)\nOutliers Removed (~26,608 rows)')
D = create_node(3, 9, node_width, node_height, 'Feature Engineering\nhumidity_binned, log_kWh, is_weekend, day_of_week_Tuesday')
E = create_node(3, 7.5, node_width, node_height, 'EDA\nACF/PACF (Lag 1–2)\nKruskal-Wallis (p = 0.0)')
F = create_node(3, 6, node_width, node_height, 'Model Exploration\nOLS (R² = 0.966), SARIMAX, XGBoost')
G = create_node(3, 4.5, node_width, node_height, 'Model Selection')
H = create_node(1.5, 3, node_width, node_height, 'Full XGBoost\nTest R²: 0.9994\nMSE: 9.04e-10')
I = create_node(4.5, 3, node_width, node_height, 'Low Humidity XGBoost\nTest R²: 0.989\nlog_kWh ~98.9%')
J = create_node(3, 1.5, node_width, node_height, 'Portfolio Visuals\nFeature Importance\nActual vs. Predicted\nPower BI (CO2 by Humidity)')
K = create_node(3, 0, node_width, node_height, 'Findings\nHigher Humidity, Lower CO2\n$16,200/year Savings')

# Draw connections
draw_arrow(A, B)
draw_arrow(B, C)
draw_arrow(C, D)
draw_arrow(D, E)
draw_arrow(E, F)
draw_arrow(F, G)
draw_arrow(G, H, 'High Performance')
draw_arrow(G, I, 'Humidity Focus')
draw_arrow(H, J)
draw_arrow(I, J)
draw_arrow(J, K)

# Set axis properties
ax.set_xlim(0, 8)
ax.set_ylim(-0.5, 14)
ax.axis('off')
plt.title('Model Creation Flowchart')

# Save the figure
plt.savefig(r'C:/Users/srsch/OneDrive/Desktop/model_flowchart_matplotlib.png', dpi=300, bbox_inches='tight')
plt.show()

print("Flowchart saved to C:/Users/srsch/OneDrive/Desktop/model_flowchart_matplotlib.png")
```
