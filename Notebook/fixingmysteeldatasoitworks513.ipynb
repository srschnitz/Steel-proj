```python
import pandas as pd

#Load my dataset with an updated date format and lets get analyzing
df = pd.read_csv('C:/Users/srsch/OneDrive/Desktop/merged_steel_weather_data_final.csv')
df.info()
df.head()
```

```python
# Create your Load_Type_Rank column
if 'Load_Type' in df.columns:
    # Mapping based on the ACTUAL values in your dataset
    load_type_mapping = {
        'Light_Load': 1,
        'Medium_Load': 2,
        'Maximum_Load': 3
        # Add more mappings if there are other load types in your data
    }
    
    # Create the rank column
    df['Load_Type_Rank'] = df['Load_Type'].map(load_type_mapping)
    
    # Check if we have any unmapped values
    nan_count = df['Load_Type_Rank'].isna().sum()
    if nan_count > 0:
        print(f"Warning: Found {nan_count} rows with unmapped Load_Type values")
        # Print the unique unmapped values to help identify what's missing
        unmapped_types = df.loc[df['Load_Type_Rank'].isna(), 'Load_Type'].unique()
        print(f"Unmapped values: {unmapped_types}")
        
        # Fill NaN values with a default value if needed
        # df['Load_Type_Rank'] = df['Load_Type_Rank'].fillna(-1)
    
    print("Created Load_Type_Rank column successfully")
else:
    print("Load_Type column not found - couldn't create Load_Type_Rank")

# Verify the mapping worked correctly
if 'Load_Type_Rank' in df.columns:
    print("\nVerification of Load_Type_Rank values:")
    print(df[['Load_Type', 'Load_Type_Rank']].drop_duplicates())
# Save to CSV
output_path = 'C:/Users/srsch/OneDrive/Desktop/merged_steel_weather_data_with_datetime.csv'
df.to_csv(output_path, index=False)
print(f"Saved data to {output_path}")

# Verify the save worked by reloading and checking
verification = pd.read_csv(output_path)
print("\nVerification after reload:")
print(f"Columns in reloaded file: {verification.columns.tolist()}")
print(f"Load_Type_Rank in columns: {'Load_Type_Rank' in verification.columns}")
print(f"Date column type after reload: {verification['date'].dtype}")

# Convert date back to datetime after loading
verification['date'] = pd.to_datetime(verification['date'])
print(f"Date column type after reconversion: {verification['date'].dtype}")
```
